```
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

**Loading of images and plotting**

```
images_directory_path = "./../images/"
file_list = os.listdir(images_directory_path)
print(file_list)
```

### **The images look unuusual because the image is read as BGR**

```
%matplotlib inline

fig = plt.figure(figsize=(8, 8))
columns = 2
rows = 2
for i in range(1, columns*rows +1):
    img1 = cv2.imread(images_directory_path + file_list[i - 1], 3)
    fig.add_subplot(rows, columns, i)
    plt.imshow(img1)
plt.show()
```

### **We convert them to RGB here**

```
%matplotlib inline

fig = plt.figure(figsize=(8, 8))
columns = 2
rows = 2
for i in range(1, columns*rows +1):
    img1 = cv2.imread(images_directory_path + file_list[i - 1], 3)
    b,g,r = cv2.split(img1)           # get b, g, r
    img1 = cv2.merge([r,g,b])     # switch it to r, g, b
    fig.add_subplot(rows, columns, i)
    plt.imshow(img1)
plt.show()
```

**Converting one colour image to grayscale. i.e Taking average of 3 channels.**

```
img = cv2.imread(images_directory_path + file_list[0], 3)
b,g,r = cv2.split(img)           # get b, g, r
img = cv2.merge([r,g,b])     # switch it to r, g, b

img = img.sum(axis=2) / 3 # Average of 3 channels
img = img.astype("int64")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

# Saving Image
cv2.imwrite(images_directory_path + "output.jpg", img)
```

**Printing dimensions of one image**

```
img = cv2.imread(images_directory_path + file_list[0], 3)
b,g,r = cv2.split(img)           # get b, g, r
img = cv2.merge([r,g,b])     # switch it to r, g, b

print("(Height, Width, Channels) = " + str(img.shape))
```

**Range of values**

The values lie in the range (0, 255), which means each value is represented using 8 bits.

```
img = cv2.imread(images_directory_path + file_list[0], 3)
b,g,r = cv2.split(img)           # get b, g, r
img = cv2.merge([r,g,b])     # switch it to r, g, b

minimum = img.min()
maximum = img.max()

print("Range = (" + str(minimum) + ", " + str(maximum) + ")")
```
